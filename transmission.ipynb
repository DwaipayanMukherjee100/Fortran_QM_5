```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
N=10000
x_min=-5
x_max=5
```

```python
def K(V,E):
    return np.sqrt(2*(E-V)+0j)
def solver(E,a):
    x=np.linspace(x_min,x_max,N)
    psi = np.zeros(N, dtype=complex)
    psid=np.zeros(N, dtype=complex)
    iota=0.0+1j 
    dx=h=x[1]-x[0]
    V = np.where(np.abs(x) < a, -10.0, 0.0)
    psi[0]  = 1e-6
    k0 = K(V[0], E)
    psid[0] = k0 * psi[0]
    for i in range(1,N):
        k1=K(V[i],E)
        psid[i]=psid[i-1]-k1*k1*psi[i-1]*dx
        psi[i]=psi[i-1]+psid[i]*dx
    psi=psi/(np.max(np.abs(psi)**2))
    psid=psid/(np.max(np.abs(psid)**2))

    plt.plot(x,np.abs(psi)**2)
    plt.title(E)
    plt.show()

for i in range(40):
    solver(0.1*i,1.0)
    
```
